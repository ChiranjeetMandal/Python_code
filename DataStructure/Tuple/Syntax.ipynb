### Creating a tuple

```python
# Creating a Tuple
t1= (1,2,3)
print(t1)
print(type(t1))

# Creating heterogo 
t2 = (1,2,3.5,"ok")
print(t2)
print(type(t2))

#Creating an empty Tuple
t3 = (1,)
print(t3)
print(type(t3))

# if we don't use comma then it will consider as integer.
t4 = (1)
print(t4)
print(type(t4))

# Creating an empty Tuple
t5 = ()
print(t5)
print(type(t5))
```

### Accessing an Element

```python
t1 = (10,44,32,66,77,3,44)
print(t1[0])
print(t1[2])
print(t1[-1])


# Slicing

print(t1[::-1])
print(t1[:2])
print(t1[3:])
```

### Operation

```python
# concatenation(+)
t1 = (2,3)
t2 = (3,4)
print(t1+t2)

# Repeatation(*)
t3 = (1,2)
print(t3 * 3)
```

### Iteration

```python
t = (1,2,3,4,5)
for i in t:
    print(i)
```

### Length

```python
t = (2,3,4,56,2,4)
print(len(t))
```

### Index

```python
t = (3,5,6,7,2,3,1)
print(t.index(6))
```

### Packing and Unpacking

```python
# Packing
t1 = 1,2,3
print(t1)

# Unpacking
t2 = (6,7,8,9)
a,b,c,d = t2
print(a)
print(b)
print(c)
print(d)
```

### Swaping Variables

```python
a = 10
b = 20

a,b = b,a
print(a)
print(b)
```

### Casting 

```python
# tuple --> list
t = (1, 2, 3)
lst = list(t)
print(lst)


# lst = [1, 2, 3]
t = tuple(lst)
print(t)
```
