# SET :
### sets are exactly like lists except for the fact that their elements are immutable (that means you cannot change/mutate an element of a set once declared). However, you can add/remove elements from the set. It is a mutable, unordered group of elements.

# Empty Set


```python
# To create an empty set always need to use set(). {} -> this will create an dictionary
set1 = set()
set2 = {}
print(type(set1))
print(type(set2))


# Set with Values

nums = {1,2,3,4,5}
print(f"Value of nums: {nums}")
print(f"Datatype of the nums: {type(nums)}")

fruits = {'Apple','Banana','Orange','Graphs'}
print(f"Value of fruits: {fruits}")
print(f"Datatype of the fruits: {type(fruits)}")

hetero = {1,"Apple",2,"Orange",3.14}
print(f"Value of hetero: {hetero}")
print(f"Datatype of the hetero: {type(hetero)}")


# From a list it removes the duplicates automatically.
list1 = [1,2,3,4,4,5,5,6,6,7,7,7,8]
s = set(list1)
print(s)

#from a String
strng_value = "Hello"
s1 = set(strng_value)
print(s1)


# From Range 
s2 = set(range(5))
print(s2)

```

### Adding and Removing the elements

```python
nums1 = {1,2,3,4,5}

#add() to add the single elements in a set
nums1.add(5)
print(nums1) # add 5 in the set
nums1.add(2)
print(nums1) # no duplicates has been added


#update() -> add multiple elements in a set
nums1.update({6,7,8})
print(nums1) # can update set dataType
nums1.update([9,10,11])
print(nums1) # can update list dataType
nums1.update({12},[13])
print(nums1) # can update both list and set


#remove() --> remove the element from the set. As set is unordered so index will not work.
nums1.remove(3)
print(nums1) # it will remove the 3 from set

#nums1.remove(99) # it will through keyError as 99 not present in the set.



#discard() -> it will remove the same as remove() does but here we won't get keyError.
nums1.discard(4)
print(nums1)
nums1.discard(99) # it will run the code 
print(nums1)


#pop() -> remove & return a random element
nums1.pop()
print(nums1)

# clear()  -> removes all the element in set
nums1.clear()
print(nums1)
```
