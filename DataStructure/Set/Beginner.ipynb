#### 1. Remove duplicates from a list
nums = [1,2,2,3,4,4,5]
Output: [1,2,3,4,5]

```python
nums = [1,2,2,3,4,4,5]

s = set(nums)
print(list(s))
```

#### 2. Count unique elements
nums = [1,1,2,3,3,4,5]

```python
nums = [1,2,2,3,4,4,5]
s1 = set(nums)
print(f"the count of unique elements: {len(s1)}")
```

#### 6. Union on multiple list
Input:  [[1,2,3],[2,3,4],[3,4,5]]

Output: [1, 2, 3, 4, 5]

```python
Input = [[1,2,3],[2,3,4],[3,4,5]]

#Method 1
op = []
for i in Input:
    for j in i:
        op.append(j)

print(list(set(op)))


#Method 2
result = set()
for lst in Input:
    result.update(lst)
print(list(result))
```
