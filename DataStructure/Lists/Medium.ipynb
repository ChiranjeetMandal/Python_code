#### Given a list of numbers and a target, return the indices of two numbers that add up to the target. Assume exactly one solution exists.

Input:  nums=[2,7,11,15], target=9
Output: [0, 1]

```python
nums = [2,7,11,15]
target = 9
for i in range(len(nums)):
    for j in range(len(nums)):
        if nums[i]+nums[j] == target:
            print([i,j])
            break
```

```python
nums = [2,7,11,15]
target = 17
for i in nums:
    rem = target - i
    if rem in nums:
        print([nums.index(i),nums.index(rem)])
        break
```

### Flatten the nested List
Input:  [[1,2], [3,4], [5]]

```python
input = [[1,2],[3,4],[5]]
output = []
for i in input:
    for j in i:
        output.append(j)
print(output)

# list comprehension
op = []
flatten =[op.append(j) for i in input for j in i]
print(op)
```

### Group Consicutive Duplicates
Input:  [1,1,2,3,3,3,4,4]

Output: [[1,1],[2],[3,3,3],[4,4]]

```python
nums = [1,1,2,3,3,3,4,4]

output = []
temp = [nums[0]]

for i in range(1, len(nums)):
    if nums[i] == nums[i-1]:
        print(f"i : {i}")
        print(f"nums[i]: {nums[i]}")
        print(f"nums[i-1]: {nums[i-1]}")
        temp.append(nums[i])
        print(f"if part:{temp}")
        
    else:
        output.append(temp)
        print(f"else part output: {output}")
        temp = [nums[i]]
        print(f"else temp assign {temp}")

output.append(temp)

print(output)
```

### Find all pairs with given sum
Input:  nums=[1,2,3,4,5], target=5

Output: [(1,4),(2,3)]

```python
nums = [1,2,3,4,5]
target = 5
new_lst = []
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        if nums[i]+nums[j] == target:
            new_lst.append([nums[i],nums[j]])

print(new_lst)
```
