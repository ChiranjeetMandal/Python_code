### Find the largest Element in a list

```python
list1 = [4,2,7,8,1,33,4,88,33,22,6,9,2,88]
print(max(list1))
```

#### Find the smallest element in a list.

```python
list1 = [4,2,7,8,1,33,4,88,33,22,6,9,2,88]
print(min(list1))
```


#### Find the sum of all elements in a list.

```python
my_list = [1,2,3,4,5,6,7,8,9]
print(sum(my_list))
```

#### Count even and odd numbers in a list.


```python
my_list = [1,2,3,4,5,6,7,8,9]
count_even = 0
count_odd = 0
for i in my_list:
    if i%2==0:
        count_even +=1
    else:
        count_odd +=1
print(f"Total count of even numbers : {count_even}")
print(f"Total count of Odd numbers : {count_odd}")
```

#### Reverse a list without using reverse().

```python
my_list1 = [1,3,5,6,7,8,9]
print(my_list1[::-1])
```

#### Remove duplicates from a list.


```python
my_list2 = [1,3,4,4,5,6,6,7,8,8,9,0]
l1 = []
for i in my_list2:
    if i not in l1:
        l1.append(i)
print(f"Unique value in a the list : {l1}")
```

#### Find the frequency of each element.

```python
#method1
my_list6 = [2,3,4,4,5,6,7,6,7,8,8,2,6]
dct = {}
for i in my_list6:
    if i not in dct:
        dct[i]= 1
    else:
        dct[i]+=1
print(dct)
```
