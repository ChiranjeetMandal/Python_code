## Dictionary : 
##### A dictionary in Python is a built-in data structure that stores data in key: value pairs, allowing you to look up information using a specific label rather than a numeric index. Dictionaries are written with curly braces {}

```python
student = {
    "name": "John",
    "age": 25,
    "city": "Hyderabad"
}
```

### Accessing an Element

```python
print(student["name"])
print(student.get("name"))
print(student.get("age"))
```

### Add or Update

```python
# update
student["age"]=26
print(student["age"])

# Add
student["stream"] = "CSE"
print(student.get("stream"))
```

### Delete an element

```python
print(student)
del student["city"]
print(student)
```

```python
print(student)
print(student.pop("city"))
print(student)

student.clear()
print(student)
```

### Get all Keys

```python
print(student.keys()) # return will in list
```

### Get all Values

```python
print(student.values()) # return will in list
```

### Get all Key-Value pairs

```python
print(student.items())
```

### Iterate through dictionary

```python
print("Method:1")
for i in student:
    print(i, student[i])


print("\nMethod:2")
for key, value in student.items():
    print(key, value)
```

### Merge Dictionary

```python
d1 = {"a": 1}
d2 = {"b": 2}

d1.update(d2)

print(d1)
# {'a': 1, 'b': 2}
```
